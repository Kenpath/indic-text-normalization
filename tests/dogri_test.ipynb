# Dogri Text Normalization Test Cases

This notebook tests the Dogri text normalization implementation.

```python
import sys
import os

# Add the NeMo-text-processing directory to the path
sys.path.insert(0, os.path.abspath('.'))

from nemo_text_processing.text_normalization.normalize import Normalizer

print("Imports successful!")
```

```python
# Initialize Dogri normalizer
normalizer_do = Normalizer(
    input_case='cased',
    lang='do',
    cache_dir=None,  # Set to a directory path if you want to cache .far files
    overwrite_cache=False,
    post_process=True
)

print("Dogri Text Normalizer initialized successfully!")
print(f"Language: {normalizer_do.lang}")
```

## 1. Cardinal Numbers Test


```python
cardinal_tests = [
    "155=100",
    "१२३४",
    "१२३४५",
    "१२३४५६",
    "१२३४५६७",
    "१२३४५६७८",
    "-१२३",
    "-120",
    "१०००",
    "१०००००",  # 1 lakh
    "१००००००००",
    "456",
    "9943206870",
    "1,234,566"
]

print("=" * 60)
print("CARDINAL NUMBERS TEST")
print("=" * 60)
for test in cardinal_tests:
    result = normalizer_do.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 1.2. Time Test (English Digits)


```python
# Test English digits for time - should work now!
time_tests_english = [
    "12:30",
    "1:40",
    "12:00",
    "12:14",
    "12:30:45",
    "9:15",
    "09:15",
    "23:59",
    "12:49",
]

print("=" * 60)
print("TIME TEST (ENGLISH DIGITS)")
print("=" * 60)
for test in time_tests_english:
    result = normalizer_do.normalize(test)
    print(f"Input:  {test:15} -> Output: {result}")
```

## 11. Mixed Content Test


```python
mixed_tests = [
    "अज्ज १५-०६-२०२४ गी १२:३० बजे मीटिंग ऐ.",
    "₹१००० ते ₹५०० जोड़ने च ₹१५०० होंदा ऐ.",
    "१२३ किलोग्राम वजन ते ५० किलोमीटर दूरी.",
    "पैहली जगह ते दूई जगह.",
    "मेरा मोबाइल नंबर 9876543210 ऐ."
]

print("=" * 60)
print("MIXED CONTENT TEST")
print("=" * 60)
for test in mixed_tests:
    result = normalizer_do.normalize(test)
    print(f"Input:  {test}")
    print(f"Output: {result}")
    print("-" * 60)
```

## 12. Batch Testing


```python
# Test multiple inputs at once
batch_tests = [
    "123",
    "१२.३४",
    "१२:३०",
    "₹१००",
    "१५-०६-२०२४",
]

print("=" * 60)
print("BATCH TESTING")
print("=" * 60)
results = normalizer_do.normalize_list(batch_tests)
for input_text, output_text in zip(batch_tests, results):
    print(f"Input:  {input_text:15} -> Output: {output_text}")
```
